```python
#Importing relevant libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
```

```python
data1 = pd.read_excel("DATA1.xlsx")
data1.head()
```

```python
data2 = pd.read_excel("DATA2.xlsx")
data2.head()
```

```python
# Rename Columns
data2.rename(columns={'Postal.Code':'Postcode', 'Location':'City'}, inplace=True)
```

```python
#function to index Address. Locator function
def getAddress(address, location):
    url = f'https://google.com/search?q={address}+{location}'
    results = requests.get(url)
    soup = bs(results.text, "html.parser")
    return soup.find("div", class_='BNeawe').text
```

```python
for point in range(len(data2)): 
    address =data2.loc[point, 'Address']
    loca = data2.loc[point, 'City']
    data2.loc[point, 'new_address'] = getAddress(address, loca)
```

```python
data2.head()
```

```python
for point in range(len(data1)): 
    address =data1.loc[point, 'Address']
    loca = data1.loc[point, 'City']
    data1.loc[point, 'new_address'] = getAddress(address, loca)
```

```python
# Combine two Data Frames using Pandas Merge
data = data1['Company.Name', 'new_address'].merge(data2, on=['new_address'], how='left')
```
